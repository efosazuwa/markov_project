```python
import pandas as pd
import cufflinks as cf
%matplotlib inline
cf.go_offline()
```

```python
t_matrix = pd.read_csv('1-gram_shakespeare_transition_matrix.csv', index_col=0)
t_matrix.head()
```

```python
fake_s_matrix = pd.read_csv('fake_shakespeare_transition_matrix.csv', index_col=0)
fake_s_matrix.head()
```

```python
t_matrix.rename(columns={' ':"'space'"}, inplace=True)
fake_s_matrix.rename(columns={' ':"'space'"}, inplace=True)

t_matrix.rename(index={' ':"'space'"}, inplace=True)
fake_s_matrix.rename(index={' ':"'space'"}, inplace=True)
```

```python
a = t_matrix.copy()
b = fake_s_matrix.copy()
```

```python
a.iplot(kind='heatmap', colorscale='blues', xTitle='Initial State', yTitle='Next State', title='Ireland Transition Matrix')
```

```python
b.iplot(kind='heatmap', colorscale='blues')
```

```python
nt_matrix=a.copy()
nfake_s_matrix=b.copy()
```

```python
from ipywidgets import interact
def plot_it2(letters):
    a = pd.concat([nt_matrix[letters], pd.DataFrame(nfake_s_matrix[letters])], axis=1)
    a.columns = ['shakespeare', 'ireland']
    a.iplot(kind='bar', title="Next letter frequency for '{}'".format(letters))
    
interact(plot_it2, letters=nt_matrix.columns.tolist())
```

```python
ax=t_matrix['a'].iplot(kind='bar')
fake_s_matrix['a'].iplot(kind='bar',ax=ax)
```
